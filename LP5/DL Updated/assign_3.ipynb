```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels=['t-shirt','trousers','pullover','dress','coat','sandal','sneakers','shirt','bag','ankle boots']
```

```python
train = pd.read_csv("fashion-mnist_train.csv")
test = pd.read_csv("fashion-mnist_test.csv")

train.head()
```

```python
x_train = train.drop(['label'],axis=1)
y_train = train['label']
```

```python
x_test = test.drop(['label'],axis=1)
y_test = test['label']
```

```python
x_train
```

```python
x_test
```

```python
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
```

```python
x_train_reshaped = x_train.values.reshape(-1,28,28,1)
x_test_reshaped = x_test.values.reshape(-1,28,28,1)
```

```python
for i in range(20):
    print(labels[y_train[i]])
    plt.imshow(x_train_reshaped[i])
    plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Flatten,Dense
```

```python
model = Sequential()
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train_reshaped,y_train,epochs=5,validation_data=(x_test_reshaped,y_test))
```

```python
loss,acc = model.evaluate(x_test_reshaped,y_test)
print(f"Loss: {loss}, Accuracy:{acc}")
```

```python
predictions = model.predict(x_test_reshaped[14:15])
```

```python
labels = labels[np.argmax(predictions)]
print(labels)
```
